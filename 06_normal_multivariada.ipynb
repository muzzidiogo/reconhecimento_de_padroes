```python
import numpy as np
import matplotlib.pyplot as plt
from utils import gen_2D_gaussians
np.random.seed(0)
```

```python
X, Y = gen_2D_gaussians()

plt.scatter(X[:100, 0], X[:100, 1], c='b', label='Class 1')
plt.scatter(X[100:, 0], X[100:, 1], c='r', label='Class -1')
plt.legend()
plt.xlim(0,3)
plt.ylim(0,3)
plt.title('Dataset sint√©tico: Gaussianas 2D. n = 100 s = 0.3')
```

```python
x1, x2 = X[:100], X[100:]
k1, k2 = np.cov(x1.T), np.cov(x2.T)
m11, m12 = np.mean(x1[:, 0]), np.mean(x1[:, 1])
m21, m22 = np.mean(x2[:, 0]), np.mean(x2[:, 1])
m1 = np.array([m11, m12])
m2 = np.array([m21, m22])
```

```python
from utils import pdfnvar

seqi = np.linspace(0, 3, 300)
seqj = np.linspace(0, 3, 300)
seq_len = len(seqi)

M1 = np.zeros((seq_len, seq_len))
M2 = np.zeros((seq_len, seq_len))
M3 = np.zeros((seq_len, seq_len))

for i in range(len(seqi)):
    for j in range(len(seqj)):
        xt = np.array([[i], [j]]) 
        M1[i, j] = pdfnvar(xt, m1, k1, 2)
        M2[i, j] = pdfnvar(xt, m2, k2, 2)
        M3[i, j] = 1 if M2[i, j] >= M1[i, j] else -1
```
