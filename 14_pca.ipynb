# Principal Component Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
from utils import make_normal_data
```

```python
np.random.seed(40)  
s = 0.6
n = 50
c1 = np.array([4, 4])
c2 = np.array([2, 2])
x1, y1 = make_normal_data(s = s, n = n, c = c1, dim = 2, label = 1)
x2, y2 = make_normal_data(s = s, n = n, c = c2, dim = 2, label = -1)
plt.scatter(x1[:, 0], x1[:, 1], c='red', label='Class 1')
plt.scatter(x2[:, 0], x2[:, 1], c='blue', label='Class 0')
plt.xlim(0, 5)
plt.ylim(0, 5)
```

```python
def myPCA(X):
    x_mean = np.mean(X, axis=0)
    X_norm = X - x_mean
    cov_matrix = np.cov(X_norm.T)
    eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
    return eig_vals, eig_vecs
```

```python
eig_vals, eig_vecs = myPCA(np.vstack((x1, x2)))
c = np.mean(np.vstack((x1, x2)), axis=0)
plt.quiver(*c, eig_vecs[0, 0], eig_vecs[1, 0], scale=3, label='PC1')
plt.quiver(*c, eig_vecs[0, 1], eig_vecs[1, 1], scale=3, label='PC2')
plt.scatter(x1[:, 0], x1[:, 1], c='red', label='Class 1')
plt.scatter(x2[:, 0], x2[:, 1], c='blue', label='Class 0')
plt.xlim(0, 6)
plt.ylim(0, 6)
```
